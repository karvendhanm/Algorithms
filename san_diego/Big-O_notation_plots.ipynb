```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = np.linspace(1, 100)
plt.plot(n, 7*(n**2)+6*n+5, label='7n^2+6n+5')
plt.plot(n, n**2, label='n^2')
plt.legend()
plt.show()
```

```python
n = np.linspace(1, 100)
plt.plot(n, 7*(n**2)+6*n+5, label='7n^2+6n+5')
plt.plot(n, 20*n, label='20*n')
plt.legend()
plt.show()
```

# Smaller terms can be omitted

```python
n = np.linspace(1, 5)
plt.plot(n, 7*(n**2)+6*n+5, label='7n^2+6n+5')
plt.plot(n, 7*(n**2), label='7n^2')
plt.legend()
plt.show()
```

```python
n = np.linspace(1, 100)
plt.plot(n, 7*(n**2)+6*n+5, label='7n^2+6n+5')
plt.plot(n, 7*(n**2), label='7n^2')
plt.legend()
plt.show()
```

```python
n = np.linspace(1, 100)
plt.plot(n, (7*(n**2)+6*n+5)/(7*(n**2)))
plt.show()
```


# Multiplicative contants can be omitted


```python
n = np.linspace(1, 100)
plt.plot(n, (7*(n**2)+6*n+5)/(n**2))
plt.show()
```

# Out of two polynomials, the one with larger degree grows faster

```python
n = np.linspace(1, 10)
plt.plot(n, n, label="n")
plt.plot(n, n * n, label="n^2")
plt.plot(n, n * n * n, label="n^3")
plt.legend(loc='upper left')
plt.show()
```

```python
n = np.linspace(1, 100)
plt.plot(n, n, label="n")
plt.plot(n, n * n, label="n^2")
plt.plot(n, n * n * n, label="n^3")
plt.legend(loc='upper left')
plt.show()
```

# Any polynomial grows slower than any exponent

```python
n = np.linspace(1, 10)
plt.plot(n, n ** 4, label="n^4")
plt.plot(n, 2 ** n, label="2^n")
plt.legend(loc='upper left')
plt.show()
```

```python
n = np.linspace(1, 20)
plt.plot(n, n ** 4, label="n^4")
plt.plot(n, 2 ** n, label="2^n")
plt.legend(loc='upper left')
plt.show()
```

# Any polylogarithm grows slower than any polynomial

```python
n = np.linspace(1, 20)
plt.plot(n, n, label="n")
plt.plot(n, np.log(n), label="log n")
plt.legend(loc='upper left')
plt.show()
```

```python
n = np.linspace(1, 100)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
plt.show()
```

```python
n = np.linspace(1, 10 ** 6)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
plt.show()
```

```python
n = np.linspace(1, 10 ** 8)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
plt.show()
```

```python
n = np.linspace(1, 10 ** 15)
plt.plot(n, n ** .5, label="n^.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
plt.show()
```

```python
n = np.linspace(1, 10**100)
plt.plot(n, n ** .1, label="n^.1")
plt.plot(n, np.log(n) ** 5, label="(log n)^5")
plt.legend(loc='upper left')
plt.show()
```
